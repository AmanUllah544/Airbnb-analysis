# checking missing values

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['last review']=pd.to_datetime(df['last review'],errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0, 'last review':df['last review'].min()}, inplace=True)
```

```python
df.head()
```

```python
df.dropna(subset=['NAME', 'host name'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['house_rules', 'license'],errors= 'ignore')
```

```python
df.head()
```

```python
# remove dollar sign from price and service fee and convert to float
df['price']=df['price'].replace('[\$]','', regex=True).str.replace(',','').str.strip().astype(float)
df['service fee']=df['service fee'].replace('[\$]','', regex=True).str.replace(',', '').str.strip().astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

# Visualization

# what is the distribution of listing price?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins= 50, kde=True, color='skyblue')
plt.title('Distribution of listing price',fontsize=30)
plt.xlabel('Price $', fontsize=30)
plt.ylabel('Frequency', fontsize=30)
plt.show()
```

# How are different room type distributed?

```python
df['room type'].unique()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='room type',data=df, hue='room type')
plt.title('Room type distribution')
plt.xlabel('room type')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(y='neighbourhood group',data=df,hue='neighbourhood group', order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by neighbourhood group')
plt.xlabel('count')
plt.ylabel('neighbourhood group')
plt.legend(loc='lower right')
plt.show()
```

# what is the relationship between price and room type

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='room type', y='price', data=df, hue='room type',palette='Set1')
plt.title('room type vs price')
plt.xlabel('room type')
plt.ylabel('price')
plt.show()
```

# How has the number of reviews change over with time?

```python
df=df[df['last review']< '2025-01-01']
review_over_time=df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12,6))
review_over_time.plot(kind='line', color='red')
plt.title('Number of review over time')
plt.xlabel('date')
plt.ylabel('number of reviews')
plt.show()
```
